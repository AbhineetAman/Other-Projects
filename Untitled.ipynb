```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_sheet1 = 'Idle' 
my_sheet2= 'peace'
my_sheet3= 'rocknmetal'
file_name = 'Sub1.xlsx' 
df_idle =pd.read_excel(file_name, sheet_name = my_sheet1,header=1)
df_peace=pd.read_excel(file_name, sheet_name = my_sheet2,header=1)
df_rocknmetal=pd.read_excel(file_name, sheet_name = my_sheet3,header=1)
```

```python
list_power=[ 'POW_Theta', 'POW_Alpha', 'POW_BetaL', 'POW_BetaH',
       'POW_Gamma','POW.F3.Alpha','POW.F4.Alpha']
```

```python
df_idle=df_idle[list_power].dropna(axis=0)
df_rocknmetal=df_rocknmetal[list_power].dropna(axis=0)
df_peace=df_peace[list_power].dropna(axis=0)
```

```python
df_idle=df_idle[list_power].loc[~(df_idle[list_power]==0).all(axis=1)]

df_rocknmetal=df_rocknmetal[list_power].loc[(df_rocknmetal[list_power]!=0).any(axis=1)]

df_peace=df_peace[list_power].loc[~(df_peace[list_power]==0).all(axis=1)]
```

```python
df_peace.shape
```

```python
df_idle=df_idle.head(48)
df_idle.reset_index(inplace = True, drop = True)
df_peace.reset_index(inplace = True, drop = True)
df_rocknmetal.reset_index(inplace = True, drop = True)
```

```python
#FAI=Frontal Asymmetry Index 
#Frontal asymmetry can be extracted from EEG headsets with electrodes located at frontal scalp regions (ideally, F3 and F4), allowing you to analyze short-term changes in motivation over the course of a stimulus presentation
df_idle['FAI']=np.log(df_idle['POW.F4.Alpha']/df_idle['POW.F3.Alpha'])
df_peace['FAI']=np.log(df_peace['POW.F4.Alpha']/df_peace['POW.F3.Alpha'])
df_rocknmetal['FAI']=np.log(df_rocknmetal['POW.F4.Alpha']/df_rocknmetal['POW.F3.Alpha'])
```

```python
df_idle['FAI'].plot()
```

```python
df_peace['FAI'].plot()
```

```python
df_rocknmetal['FAI'].plot()
```
